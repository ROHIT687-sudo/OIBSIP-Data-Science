```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# Load both of your files
df_india = pd.read_csv("Unemployment in India.csv")
df_rate = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

# Clean column names by stripping trailing hidden whitespaces
df_india.columns = df_india.columns.str.strip()
df_rate.columns = df_rate.columns.str.strip()

print("File 1 Columns:", df_india.columns.tolist())
print("File 2 Columns:", df_rate.columns.tolist())
```

```python
# Remove any empty rows if they exist
df_india = df_india.dropna()
df_rate = df_rate.dropna()

# Standardize column naming for easy analysis
df_india.rename(columns={'Estimated Unemployment Rate (%)': 'Unemployment_Rate', 
                        'Estimated Employed': 'Employed', 
                        'Estimated Labour Participation Rate (%)': 'Labour_Rate'}, inplace=True)

df_rate.rename(columns={'Estimated Unemployment Rate (%)': 'Unemployment_Rate', 
                       'Estimated Employed': 'Employed', 
                       'Estimated Labour Participation Rate (%)': 'Labour_Rate'}, inplace=True)

# Convert Date strings into actual Python datetime objects
df_india['Date'] = pd.to_datetime(df_india['Date'].str.strip(), dayfirst=True)
df_rate['Date'] = pd.to_datetime(df_rate['Date'].str.strip(), dayfirst=True)

print("Data Cleaning Complete!")
```

```python
fig_timeline = px.line(df_rate, x='Date', y='Unemployment_Rate', color='Region',
                       title='Unemployment Rate Trends Across Regions (Covid-19 Era)',
                       labels={'Unemployment_Rate': 'Unemployment Rate (%)'})
fig_timeline.update_xaxes(rangeslider_visible=True)
fig_timeline.show()
```

```python
# Group data by state region
regional_avg = df_rate.groupby('Region')['Unemployment_Rate'].mean().reset_index()
regional_avg = regional_avg.sort_values(by='Unemployment_Rate', ascending=False)

fig_bar = px.bar(regional_avg, x='Region', y='Unemployment_Rate',
                 title='Average Unemployment Rate by State/Region (Jan-Nov 2020)',
                 color='Unemployment_Rate',
                 labels={'Unemployment_Rate': 'Average Rate (%)'})
fig_bar.show()
```

```python
fig_box = px.box(df_india, x='Area', y='Unemployment_Rate', color='Area',
                 title='Unemployment Impact: Urban Sectors vs Rural Regions',
                 labels={'Unemployment_Rate': 'Unemployment Rate (%)', 'Area': 'Location Type'})
fig_box.show()
```

```python
fig_sun = px.sunburst(df_india, path=['Region', 'Area'], values='Unemployment_Rate',
                      title='Regional & Sector-wise Distribution of Unemployment')
fig_sun.show()
```
